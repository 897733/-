# проанализируем, как количество спален влияет на стоимость недвижимости

```python
data = df['bedrooms'].value_counts() # подготовка данных( смотрим)
data
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['bedrooms'], y=df['price'])
plt.title('Влияние количества спален на цену недвижимости')
plt.xlabel('Количество спален')
plt.ylabel('Цена ($)')
plt.show()
```

## Вывод: на диаграмме видно, что с увеличением количества спален средняя цена недвижимости увеличивается, но не линейно. Среди домов с малым количеством спален (1-2 спальни) также встречаются дорогие объекты, что может указывать на влияние других факторов.

# изучим влияния жилой площади (sqft_living) на стоимость недвижимости;

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['sqft_living'], y=df['price'], alpha=0.5)
plt.title('Влияние жилой площади на цену недвижимости')
plt.xlabel('Жилая площадь (кв. футы)')
plt.ylabel('Цена ($)')
plt.show()
```
